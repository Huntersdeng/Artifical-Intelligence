```python
def joinFactors(Factor1, Factor2):
    # your code
    if (not Factor1.empty) and (not Factor2.empty):
        overlap = list(set(Factor1.columns).intersection(set(Factor2.columns)))
        overlap.remove('probs')
        if len(overlap)==0:
            return pd.merge(Factor1, Factor2)
        if len(Factor1.columns)<=len(Factor2.columns):
            biggerFactor = pd.DataFrame.copy(Factor2)
            smallerFactor = Factor1
        else:
            biggerFactor = pd.DataFrame.copy(Factor1)
            smallerFactor = Factor2
        mask = smallerFactor.drop(columns='probs')
        Factor = pd.merge(biggerFactor, mask, on=overlap, how='outer')
        elements = list(mask.columns)
        func(Factor, smallerFactor, elements, {})
        return Factor
    else:
        if Factor1.empty:
            return Factor2
        if Factor2.empty:
            return Factor1
```

```python
def func(Factor, mask, elements, dic):
    element = elements.pop()
    outcome = set(Factor[element])
    for i in outcome:
        dic[element]=i
        if len(elements)==0:
            indexes = list(dic.items())
            Factor[Factor[element]==i]['probs']*=func2(mask, indexes)
        else:
            func(Factor[Factor[element]==i], mask[element], elements, dic)
```

```python
def func2(mask, indexes):
    index = indexes.pop()
    if len(indexes)==0:
        return mask[mask[index[0]]==index[1]]['prob']
    else:
        func2(mask[mask[index[0]]==index[1]], indexes)
```
